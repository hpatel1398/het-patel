```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
overall=pd.read_csv("/content/Data/AllContentData.csv")
overall#.head()
```

```python
shows=overall[overall['object_type']=='show']
show_season=overall[overall['object_type']=='show_season']
movies=overall[overall['object_type']=='movie']
```

```python
sns.set(style="darkgrid")
ax = sns.countplot(x="object_type", data=overall, palette="Set2")
```

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="tmdb_popularity", data=movies, palette="Set2", order=movies['tmdb_popularity'].value_counts().index[0:15])
```

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="original_release_year", data=movies, palette="Set2", order=movies['original_release_year'].value_counts().index[0:15])
```

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="original_release_year", data=shows, palette="Set2", order=shows['original_release_year'].value_counts().index[0:15])
```

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="original_release_year", data=show_season, palette="Set2", order=show_season['original_release_year'].value_counts().index[0:15])
```
